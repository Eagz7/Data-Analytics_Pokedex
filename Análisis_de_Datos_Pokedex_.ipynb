```python
df_pokedex["total_stats"]=df_pokedex.get(["hp", "attack", "defense", "sp_attack", "sp_defense", "speed"]).sum(axis=1)
```

```python
df_pokedex.head()
```

```python
df_pokedex["generation"].unique()
```

```python
def generacionones(numero):

  if numero==1:

    return "Kanto"
  
  elif numero ==2:

    return "Johto"

  elif numero==3:

    return "Hoenn"

  elif numero==4:

    return "Sinnoh"

  elif numero==5:

    return "Unova"

  elif numero==6:
    
    return "Kalos"

  elif numero== 7:

    return "Alola"

```

```python
df_pokedex["region"]= df_pokedex["generation"].apply(generacionones)
```
